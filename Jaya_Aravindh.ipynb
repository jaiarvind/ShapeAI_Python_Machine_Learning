```python
#Importing Libraries
import numpy as np
import pandas as pd
import sklearn 
```

```python
#To import Dataset from Skilearn
from sklearn.datasets import load_boston
df=load_boston()
```

```python
#Creating Panda's DataFrame
boston = pd.DataFrame(df.data, columns=df.feature_names)
boston.head(-10)
```

```python
boston['MEDV']= df.target
boston.head(-10)
```

```python
#Creating Training and Testing Datasets
from sklearn.model_selection import train_test_split
X= boston.drop('MEDV', axis=1)
Y= boston['MEDV']
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size= 0.10, random_state= 4)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X= boston.drop('MEDV', axis=1)
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size= 0.10, random_state= 5)

```

```python
#To Fit model into given Dataset
lr_model= LinearRegression()
lr_model.fit(X_train, Y_train)
```

```python
Y_train_predict=lr_model.predict(X_train)
RMS=(np.sqrt(mean_squared_error(Y_train, Y_train_predict)))

print('The model performance for training set')
print('RMSE is {}', format(RMS))
print('\n')

#for Testing Set
Y_test_predict= lr_model.predict(X_test)
RMS=(np.sqrt(mean_squared_error(Y_test, Y_test_predict)))

#For training Set
print('The model performance for testing set')
print('RMSE is {}', format(RMS))
print('\n')
```
